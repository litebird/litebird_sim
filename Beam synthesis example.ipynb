```python
from litebird_sim import gauss_beam_to_alm
import healpy
import numpy as np
```

```python
alm = gauss_beam_to_alm(
    lmax=150,
    mmax=150,
    fwhm_min_rad=np.deg2rad(2.0),
    fwhm_max_rad=np.deg2rad(4.0),
    psi_ell_rad=np.deg2rad(30.0),
    psi_pol_rad=np.deg2rad(0.0),
    cross_polar_leakage=1e-7,
)
healpy.mollview(healpy.alm2map(alm, 512)[0], rot=(0.0, 90.0, 0.0))
```
